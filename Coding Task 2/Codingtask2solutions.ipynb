```python
from PIL import Image
import pillow_heif
```

```python
def convert_to_png(url_name, url_raw, url_png):
  heif_file = pillow_heif.read(url_raw+url_name)
  image = Image.frombytes(heif_file.mode,heif_file.size,heif_file.data,"raw",)
  url_name = url_name.split('.')[0]+'.png'
  image.save(url_png+url_name, format="png")

url_raw  = '/content/drive/MyDrive/Data Science Repo/Coding Task 2/raw/'
url_png = '/content/drive/MyDrive/Data Science Repo/Coding Task 2/Png_images/'
for i in range(79,94):
  url_name = 'IMG_53'+str(i)+'.HEIC'
  convert_to_png(url_name, url_raw, url_png)
```

# 2. Filtering and Thresholding

```python
def black_white(url_png, url_name, url_black_white, threshold):  
  #load image and show the B channel
  imageRGB  = io.imread(url_png+url_name)
  image = imageRGB[:,:,2]#all columns and all rows but only the third layer (2)
  image_mean = skf.median(image,selem=skm.disk(20))
  image_threshold = image_mean<threshold
  image = Image.fromarray(image_threshold)

  image.save(url_black_white+url_name, format="png")

threshold = 60
url_black_white  ='/content/drive/MyDrive/Data Science Repo/Coding Task 2/black_white/'

for i in range(79, 94):
  url_name = 'IMG_53'+str(i)+'.png'
  black_white(url_png, url_name, url_black_white, threshold)
```
